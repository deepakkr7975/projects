```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x=np.random.randint(1,20,100)
y1=x+5
y2=x**2
y3=x**3
```

```python
#scatter plot
plt.figure(figsize=(10,3))
plt.scatter(x,y1,c='g')
plt.scatter(x,y2,c='r')
plt.scatter(x,y3,c='b')
plt.title('scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
```

```python
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.scatter(x,y1,marker='*',c='g',s=100)
```

```python
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.scatter(x,y2,marker='.',c='r',s=200)
plt.show()
```

```python
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.scatter(x,y3,marker='^',c='b',s=20)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.plot(x,y2,color='r',linestyle='-',linewidth=2)

plt.plot(x,y3,color='b',linestyle='dashed',linewidth=2)
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
```

```python
plt.figure(figsize=(10,2))
plt.subplot(1,3,1)
plt.plot(x,y1,c='g',linestyle=':',linewidth=2)
plt.title('x vs y1')
```

```python
#bar plot
student= {'bob':87,'matt':56,'sam':27}
names=list(student.keys())
values=list(student.values())
plt.figure(figsize=(10,2))
plt.bar(names,values,color=['r','g','b'])
plt.title('bar plot')
plt.xticks(rotation=45)
plt.yticks(np.arange(0,101,10))
```

```python
data=pd.read_csv('/content/12. Seaborn_Batsman.csv')
```

```python
data
```

```python
from matplotlib import pyplot as plt
data.plot(kind='scatter', x='RunsScored', y='BallsFaced', s=32)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
balls_faced_by_position = data.groupby('BattingPosition')['BallsFaced'].sum()

max_position = balls_faced_by_position.idxmax()
max_balls = balls_faced_by_position.max()

print(f"Position {max_position} faced the most balls: {max_balls}")
```

```python
balls_faced_by_position = data.groupby('BattingPosition')['RunsScored'].sum()

# Find the position with the maximum balls faced
max_position = balls_faced_by_position.idxmax()
max_balls = balls_faced_by_position.max()

print(f"Position {max_position} scored the most runs: {max_balls}")
```
